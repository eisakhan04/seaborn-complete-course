# what is Seaborn Histogram / Seaborn Distplot

Seaborn offers two main functions for visualizing the distribution of data: sns.histplot() and sns.distplot(). These functions are used to create histograms and distribution plots, respectively. Here's an explanation of each:

**Seaborn Histogram (sns.histplot()):**

A histogram is a graphical representation of the distribution of a dataset. It divides the data into bins or intervals and displays the frequency or count of data points within each bin.
Seaborn's **sns.histplot() function** is designed specifically for creating histograms. It allows you to visualize the distribution of a single variable.
You can customize the number of bins, color, transparency, and other aesthetic properties of the histogram to tailor it to your needs.
Histograms are useful for understanding the shape, central tendency, and spread of a dataset. They can reveal patterns such as whether the data is normally distributed, skewed, or has multiple modes.
Seaborn Distribution Plot **(sns.distplot()):**

A distribution plot is a more versatile plot that combines a histogram with other elements, such as a kernel density estimate (KDE) or a fitted probability distribution. This provides a more detailed view of the data's distribution.
Seaborn's **sns.distplot()** function can create distribution plots. It allows you to visualize the distribution of a single variable and overlay a KDE curve or fit a parametric distribution (like a normal distribution) to the data.
Distribution plots are helpful for gaining insights into the data's distribution characteristics and for comparing it to theoretical distributions.
The sns.distplot() function has been deprecated in newer Seaborn versions in favor of sns.histplot() and sns.kdeplot(), which provide more control and flexibility over the components of the distribution plot.
In summary, Seaborn provides functions like **sns.histplot()** and **sns.distplot()** to create histograms and distribution plots, respectively, allowing you to explore and visualize the distribution of your data in different ways. These plots are valuable tools for understanding the underlying patterns and characteristics of your datasets.

```python
# Modify histogram with bins  
bins = [1,5,10,15,20,25,30,35,40,45,50,55] # list

plt.figure(figsize=(16,9))
sns.set()

sns.distplot(df["total_bill"], bins = bins)

plt.xticks(bins) # set bins value

plt.title("Histogram of Total Bill") 
plt.show()
```

The provided code modifies a histogram using Seaborn to visualize the distribution of a dataset, specifically the "total_bill" column from the DataFrame df. Here are the key aspects of the code:

**Bins Configuration:** It defines a custom list of bin edges named bins. These bin edges determine the intervals into which the data will be grouped for the histogram. Each bin represents a range of values.

**Plot Size and Style:** It sets the figure size to (16, 9) using plt.figure(figsize=(16, 9)) to adjust the dimensions of the plot. Additionally, sns.set() is called to apply the default Seaborn style to the plot.

**Seaborn distplot():** It uses Seaborn's sns.distplot() function to create the histogram. The df["total_bill"] column is passed as the data to be visualized, and the specified bins list is used to determine the bin edges.

**X-axis Ticks:** plt.xticks(bins) is used to set custom tick positions on the x-axis. This ensures that the tick marks correspond to the bin edges defined in the bins list.

**Plot Title:** It sets the title of the histogram to "Histogram of Total Bill" using plt.title().

**Display Plot:** Finally, plt.show() is called to display the modified histogram.

In summary, this code customizes a histogram's bin edges using the provided bins list, adjusts the plot size and style, and sets custom x-axis tick positions for better control and visualization of the "total_bill" distribution.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

plt.figure(figsize=(16, 9))
sns.set()

bins = [1, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55]

# Load your data into the "df" DataFrame before using it in this code

sns.distplot(df["total_bill"], bins=bins,
             hist_kws={'color': '#DC143C', 'edgecolor': '#aaff00',
                       'linewidth': 5, 'linestyle': '--', 'alpha': 0.9},
             kde=False,
             fit=norm,
             fit_kws={'color': '#8e00ce',
                      'linewidth': 2, 'linestyle': '--', 'alpha': 0.4},  # Fixed typo in linewidth
             rug=True,
             rug_kws={'color': '#0426d0', 'edgecolor': '#00dbff',
                      'linewidth': 3, 'linestyle': '--', 'alpha': 0.9},
             label="TB")

plt.xticks(bins)
plt.title("Histogram of Restaurant Total Bill", fontsize=20)  # Fixed typo in "Restorant"
plt.xlabel("Total Bill", fontsize=15)
plt.legend()
plt.show()
```
