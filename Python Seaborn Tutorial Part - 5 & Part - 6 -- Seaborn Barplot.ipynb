# What is Seaborn Barplot

A Seaborn barplot is a data visualization technique used to display the relationship between a categorical variable and a numerical variable. It represents the categorical data as rectangular bars with lengths proportional to the values they represent. Seaborn provides the sns.barplot() function to create these barplots efficiently.

**Here are the key characteristics and components of a Seaborn barplot:**

**Categorical Variable:** The x-axis of a Seaborn barplot typically represents the categories or groups of the categorical variable you want to analyze.

**Numerical Variable:** The y-axis represents the numerical variable, and the height of each bar corresponds to the values of this variable for the respective category.

**Bar Length:** The length of each bar is proportional to the values of the numerical variable, allowing you to compare the magnitudes of these values across categories.

**Error Bars (Optional):** Seaborn can also display error bars on the bars to represent confidence intervals or variability in the data, providing additional information about the precision of the measurements.

**Grouping Data:** Seaborn's sns.barplot() function can group data within categories and display the mean or other summary statistics, making it useful for comparing data across different groups.

**Customization:** Seaborn allows you to customize various aspects of the barplot, such as color, orientation, order of categories, and error bar representation, to tailor the visualization to your specific needs.

Seaborn barplots are commonly used for tasks like visualizing group-wise statistics, comparing data across categories, and identifying patterns or trends within categorical data. They are a valuable tool for data exploration and communication of insights.

# How to draw Sebaorn Barplot / Bar Graph?

Parameters
----------
x, y, hue : names of variables in ``data`` or vector data, optional
    Inputs for plotting long-form data. See examples for interpretation.        
data : DataFrame, array, or list of arrays, optional
    Dataset for plotting. If ``x`` and ``y`` are absent, this is
    interpreted as wide-form. Otherwise it is expected to be long-form.    
order, hue_order : lists of strings, optional
    Order to plot the categorical levels in, otherwise the levels are
    inferred from the data objects.        
estimator : callable that maps vector -> scalar, optional
    Statistical function to estimate within each categorical bin.
ci : float or "sd" or None, optional
    Size of confidence intervals to draw around estimated values.  If
    "sd", skip bootstrapping and draw the standard deviation of the
    observations. If ``None``, no bootstrapping will be performed, and
    error bars will not be drawn.
n_boot : int, optional
    Number of bootstrap iterations to use when computing confidence
    intervals.
units : name of variable in ``data`` or vector data, optional
    Identifier of sampling units, which will be used to perform a
    multilevel bootstrap and account for repeated measures design.    
orient : "v" | "h", optional
    Orientation of the plot (vertical or horizontal). This is usually
    inferred from the dtype of the input variables, but can be used to
    specify when the "categorical" variable is a numeric or when plotting
    wide-form data.    
color : matplotlib color, optional
    Color for all of the elements, or seed for a gradient palette.    
palette : palette name, list, or dict, optional
    Colors to use for the different levels of the ``hue`` variable. Should
    be something that can be interpreted by :func:`color_palette`, or a
    dictionary mapping hue levels to matplotlib colors.    
saturation : float, optional
    Proportion of the original saturation to draw colors at. Large patches
    often look better with slightly desaturated colors, but set this to
    ``1`` if you want the plot colors to perfectly match the input color
    spec.    
errcolor : matplotlib color
    Color for the lines that represent the confidence interval.
errwidth : float, optional
    Thickness of error bar lines (and caps).         
capsize : float, optional
    Width of the "caps" on error bars.

dodge : bool, optional
    When hue nesting is used, whether elements should be shifted along the
    categorical axis.    
ax : matplotlib Axes, optional
    Axes object to draw the plot onto, otherwise uses the current Axes.    
kwargs : key, value mappings
    Other keyword arguments are passed through to ``plt.bar`` at draw
    time.


```python
# Plot tips_df.total_bill barplot
sns.barplot(x = tips_df.total_bill)
```

```python
# Plot tips_df.day & tips_df.total_bill barplot
sns.barplot(x = tips_df.day, y = tips_df.total_bill)
```

```python
# Devide barplot using hue parameter
sns.barplot(x = tips_df.day, y = tips_df.total_bill, hue = tips_df.sex)
```

```python
# Pass dataset using data parameter
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df)
```

```python
# modify the order of day
order = ['Sun', 'Thur', 'Fri', 'Sat']

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, order = order)
```

```python

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, order = ['Sun', 'Thur', 'Fri', 'Sat'])
```

```python
#Modify hue order
hue_order = ['Female', 'Male']


sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, hue_order = hue_order)
```

```python
# estimate y variable value and then plot
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, estimator= np.mean)
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, estimator= np.max)
```

```python
# estimate y variable value and then plot
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, estimator= np.min)
```

```python
# set error bar size
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, ci = 30)
```

```python
# set error bar 
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df,  n_boot=8)
```

```python
# Set barplot as vertical
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, orient='v')
```

```python
# set barplot as horizontal
sns.barplot(y = 'day', x = 'total_bill', hue = 'sex',
           data = tips_df,)
```

```python
# set barplot as horizontal
sns.barplot(x ='total_bill', y = 'size', hue = 'sex', data = tips_df,
           orient='h')
```

```python
# set color
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, color="g")
```

```python
# set color map or palette
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, palette="magma")
```

```python
# set saturation of barplot
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, saturation=.3)
```

```python
#set color of error bar
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, errcolor='0.5')
```

```python
# set width of error bar
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, errwidth= 12)
```

```python
# set cap size of error bar
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, capsize=1)
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df,)
```

```python
# shift hue categorical variable bar bar in one bar
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, dodge= False)
```

```python
# Keyword Arguments parameter
kwargs = {'alpha':0.9, 'linestyle':'--', 'linewidth':5, 'edgecolor':'k'}

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df,**kwargs)
```

```python
# set background style darkgrid
sns.set()
kwargs = {'alpha':0.9, 'linestyle':':', 'linewidth':5, 'edgecolor':'k'}

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', 
           data = tips_df,**kwargs)
```

```python
# Pass Keyword  argument as parameter
sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, alpha =.9, linestyle = "-.", linewidth = 3,
           edgecolor = "g")
```

```python
# Axes parameter
ax = sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, alpha =.9, linestyle = "-.", linewidth = 3,
           edgecolor = "g")

ax.set(title = "Barplot of Tips DataFrame",
      xlabel = "Days",
      ylabel = "Total Bill")
```

```python
# Example of Seaborn Barplot
plt.figure(figsize = (16,9))

sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, alpha =1, linestyle = "-.", linewidth = 3,
           edgecolor = "k")

plt.title("Barplot of Days and Total Bill", fontsize = 20)
plt.xlabel("Days", fontsize = 15)
plt.ylabel("Total Bill", fontsize = 15)

plt.savefig("Barplot of Days and Total Bill")
plt.show()
```
